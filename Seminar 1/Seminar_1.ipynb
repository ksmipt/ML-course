# Python vs C++:
## Интерпретируемость vs Компилируемость
- В компилируемых языках код программы нужно скомпилировать в т.н. исполняемый код. Этот код затем будет исполнять нечто, что ничего не знает про исходный язык программирования.

- В интерпретируемом же языке исполнителем является сама программа, которая анализирует исходный код. Python - интерпретируемый язык (и исполнителем кода на нём является программа под названием "python"). Из этого следует, что большинство ошибок возникает на этапе выполнения программы.

## Типизация 
- Сильная / слабая типизация (также иногда говорят строгая / нестрогая). Сильная типизация выделяется тем, что язык не позволяет смешивать в выражениях различные типы и не выполняет автоматические неявные преобразования, например нельзя вычесть из строки множество. Языки со слабой типизацией выполняют множество неявных преобразований автоматически, даже если может произойти потеря точности или преобразование неоднозначно.

**Примеры**:    
Статическая: C, Java, C#;    
Динамическая: Python, JavaScript, Ruby.

- Статическая / динамическая типизация. Статическая определяется тем, что конечные типы переменных и функций устанавливаются на этапе компиляции. Т.е. уже компилятор на 100% уверен, какой тип где находится. В динамической типизации все типы выясняются уже во время выполнения программы. 

**Примеры**:   
Сильная: Java, Python, Haskell, Lisp;   
Слабая: C, JavaScript, Visual Basic, PHP.

- Явная / неявная типизация. Явно-типизированные языки отличаются тем, что тип новых переменных / функций / их аргументов нужно задавать явно. Соответственно языки с неявной типизацией перекладывают эту задачу на компилятор / интерпретатор.

**Примеры**:   
Явная: C++, D, C#;   
Неявная: PHP, Lua, JavaScript.

# 3. Численные методы и основы оптимизации

## Производная функции одной переменной

$f(x) = 3x^4 - 7x^2$ 

$f'(x) = 12x^3 - 14x \Rightarrow f'(1) = 12 - 14 = -2$

```python
def f(x):
    return 3 * x**4 - 7 * x**2
```

```python
def df(x_0, dx = 0.0001):
    return (f(x_0 + dx) - f(x_0)) / dx

print(df(1))
```

```python
print(df(1, 0.1))
print(df(1, 0.01))
print(df(1, 0.00000001))
print(df(1, 0.000000001))
print(df(1, 0.000000000000001))
print(df(1, 0.0000000000000001))
```

Для наглядности построим саму функцию $f(x)$ и касательную к ней в точке 1.

Уравнение касательной к графику функции $f(x)$ в точке $x_0$: $y_{tan} = f(x_0) + f'(x_0)(x - x_0)$.

```python
def y_tangent_line(x, x_0):
    return f(x_0) + df(x_0) * (x - x_0)
```

```python
import matplotlib.pylab as plt

x = np.linspace(-2, 2, num = 100)

y = f(x)
tang_line = y_tangent_line(x, 1)

plt.plot(x, y,  label = 'f(x)')
plt.plot(x, tang_line, label = 'Tangent Line')
plt.legend()
```

# Частные производные и градиент

$g(x, y) = 3x^2 + xy + 2y^2 - x - 4y$

$\textbf{grad} \hspace{3px} g(x, y) = \begin{pmatrix}
  6x + y - 1 \\
  x + 4y -4
\end{pmatrix} \Rightarrow
\textbf{grad} \hspace{3px} g(1, 1) = \begin{pmatrix}
  6 \\
  1
\end{pmatrix}$

```python
def g(x, y):
    return 3 * x**2 + x * y + 2 * y ** 2 - x - 4 * y

def grad_g(x_0, y_0, d = 0.0001):
    ans = []
    ans.append((g(x_0 + d , y_0) - g(x_0 , y_0)) / d)
    ans.append((g(x_0 , y_0 + d) - g(x_0 , y_0)) / d)
    return np.array(ans)

print(grad_g(1, 1))
```

$\textbf{Задача:}$ Доказать, что длина трехмерного вектора $v$ минимальна тогда, 
когда все компоненты этого вектора $v$ нулевые.

```python
def step(v, directoin, step_size):
    '''Двигаться в направлении direction с шагом step_size от v'''
    return [v_i + step_size * direction_i for v_i, direction_i in zip(v, directoin)]

help(step)
```

```python
def grad_sum_of_squares(v):
    return [2 * v_i for v_i in v]

def distance(v1, v2):
    return sum((v1_i - v2_i)**2 for v1_i, v2_i in zip(v1, v2)) ** (0.5)
```

```python
# Начальный вектор
v = [np.random.randint(-10, 10) for i in range(3)] 

# Константа точности
tolerance = 0.00001
```

```python
gradient = grad_sum_of_squares(v)
next_v = step(v, gradient, -0.01)

print(gradient)
print(next_v)
```

```python
while True:
    gradient = grad_sum_of_squares(v)
    next_v = step(v, gradient, -0.01)
    if distance(next_v, v) < tolerance:
        break
    v = next_v
    
print(v)
```

# 4. Парная линейная регрессия

```python
np.random.seed(108) # Для воспроизводимости результатов

size = 100
X = np.linspace(-1, 5, num = size)

def y(x):
    return 0.4 * x + 3

def generate_sample(size_, X_):
    return y(X_) + np.random.normal(0, 0.5, size = size_)


Y = generate_sample(size, X)
Y_real = y(X)

plt.plot(X, Y, 'o')
#plt.plot(X, Y_real, 'green', linewidth = 1.2)
plt.show()

print(X.shape, Y.shape)
```

```python
# Хотим восстановить уравнение прямой y(x) по точкам

print(X.mean(), Y.mean(), Y_real.mean())

print(X*Y)

# Решаем аналитически
k = ((X*Y).mean() - X.mean() * Y.mean()) / ((X*X).mean() - X.mean() ** 2)
b = Y.mean() - k * X.mean()

def ans_lin_1(k_, b_, x):
    return k_ * x + b_


ans = ans_lin_1(k, b, X)
plt.plot(X, Y, 'o')
plt.plot(X, Y_real, 'green', linewidth = 1.2)
plt.plot(X, ans)
plt.show()
```

```python
# Решим эту же задачу методом градиентного спуска
# Минимизируем сумму квадратов отклонений точек от прямой

import tqdm
import math 
import time

def ans_lin_2(x_, beta_):
    k = beta_[0]
    b = beta_[1]
    return k * x_ + b

def er(x_, y_, beta_):
    return (1 / x_.shape[0]) * sum((y_ - ans_lin_2(x_, beta_))**2)


def grad_er(x_, y_, beta_):
    k = beta_[0]
    b = beta_[1]
    B = b * np.ones(x_.shape[0], int)
    return (-2 / x_.shape[0]) * np.array([sum((y_ - k * x_ - b) * x_), sum((y_ - k * x_ - b))])

beta = np.random.normal(0, 0.5, size = 2) * math.sqrt(2/3)

print(er(X, Y, beta))
print(er(X, Y_real, beta))
```

```python
# Применяем метод градиентного спуска в задаче линейной регрессии

beta = np.random.normal(0, 0.5, size = 2) * math.sqrt(2/3)

time.start = time.time()

lam = 0.01
for epoch in tqdm.tqdm_notebook(range(1000), desc = 'Progress bar'):
    beta -= lam * grad_er(X, Y, beta)

print(time.time() - time.start, " seconds")
print(beta)
```

## Полиномиальная регрессия

На практике часто бывает так, что зависимость в данных имеет нелинейную природу. В таких случаях решением может быть линейная комбинация нелинейных признаков:
$$
    y = w_0 + w_1 \phi_1(\vec{x}) + w_2 \phi_2(\vec{x})
$$

В простейшем случае функцией $\phi$ может быть возведение в $i$-ю степень. Тогда мы получаем так называемую полиномиальню регрессию
$$
    y = w_0 + w_1 x^1 + w_2 x^2 + w_3 x^3 + ... + w_p x^p
$$

Давайте попробуем при помощи полиномиальной регрессии востановить функцию синуса.

```python
plt.figure(figsize = (10,5))

x = np.linspace(0, 2 * 3.14, num = 100)
y = np.sin(x) + np.random.normal(0, 0.2, size = x.shape[0])
plt.plot(x, y, 'o', label = 'sample')
#plt.plot(x, np.sin(x), '--', label = 'sin(x)')
plt.legend()
```

```python
plt.figure(figsize = (15,10)) # Размеры грфика

def polinom(x, deg):
    return np.array( [np.ones(x.shape[0])] + list(map(lambda n: x**n, range(1, deg + 1))) ).T
    # Функциональное программирование (см. ссылку в конце ноутбука)
    # Распространненая конструкция list(map(lambda ....))
    

for degree in [1, 4, 8, 11]:
    # Решаем задачу аналитически
    X = polinom(x, degree)
    w = np.dot(np.dot(np.linalg.inv(np.dot(X.T, X)), X.T), y)
    
    # Визуализируем результат
    y_hat = np.dot(w, X.T)
    plt.plot(x, y_hat, label = 'degree %s' % degree, linewidth = 2.2)
    print('Degree: %s, Loss: %s' % (degree, np.linalg.norm(y - y_hat)))
    
plt.plot(x, y, 'o', label = 'sample')
plt.plot(x, np.sin(x), '--', label = 'manifold')
plt.legend()
```

Визуально видно, что полином 4ой степени неплохо приближает реальную зависимость. При этом увеличение степени полинома до 8 и 11 еще больше уменьшает loss на трейне. Но визуально мы видим, что наш результат стал хуже. Данное явление называется переобучение.

Для контроля за переобучением данные случайным образом разделяют на 2 части: обучающую и тестовую выборки.

```python
from sklearn.model_selection import train_test_split

plt.figure(figsize = (15,10))

# Делим данные на обучающую и тестовую выборки
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 10)
order = np.argsort(x_train)

train_loss = []
test_loss = []

for degree in [1, 4, 8, 11]:
    # Решаем задачу аналитически
    X = polinom(x_train, degree)
    w = np.dot(np.dot(np.linalg.inv(np.dot(X.T, X)), X.T), y_train)
    
    # Визуализируем результат
    y_hat = np.dot(w, X.T)
    X_test = polinom(x_test, degree)
    y_hat_test = np.dot(w, X_test.T)
    train_l = np.linalg.norm(y_train - y_hat, 2)
    test_l = np.linalg.norm(y_test - y_hat_test)
    
    print('Degree: %s, train loss: %s, test loss: %s' % (degree, train_l , test_l ))
    plt.plot(x_train[order], y_hat[order], label = 'degree %s' % degree, linewidth = 2.2)
        
    
plt.plot(x, y, 'o', label = 'sample')
plt.plot(x, np.sin(x), '--', label = 'manifold')
plt.legend()
```

## Переобучение
Переобучение (переподгонка, англ. overfitting) в машинном обучении и статистике — явление, когда построенная модель хорошо объясняет примеры из обучающей выборки, но относительно плохо работает на примерах, не участвовавших в обучении (на примерах из тестовой выборки).

<img src="images/Overfitted_Data.png" alt="" style="width: 600px;"/>
Сигнал с шумом (близкий к линейному) аппроксимируется линейной функцией и полиномом. Хотя полином гарантирует идеальное совпадение, линейная аппроксимация лучше генерализирует закономерность и будет давать лучшие предсказания.

**Методы предотвращения переобучения:**
- Перекрёстная проверка (кросс-валидация, кросс-энтропия)
- Регуляризация (математика)

## Перекрёстная проверка
Перекрёстная проверка (кросс-проверка, скользящий контроль, англ. cross-validation) — метод оценки аналитической модели и её поведения на независимых данных. При оценке модели имеющиеся в наличии данные разбиваются на k частей. Затем на k−1 частях данных производится обучение модели, а оставшаяся часть данных используется для тестирования. Процедура повторяется k раз; в итоге каждая из k частей данных используется для тестирования. В результате получается оценка эффективности выбранной модели с наиболее равномерным использованием имеющихся данных.

## Регуляризация
**Регуляризация** в статистике, машинном обучении, теории обратных задач — метод добавления некоторых дополнительных ограничений к условию с целью решить некорректно поставленную задачу или предотвратить переобучение. Эта информация часто имеет вид штрафа за сложность модели.

**Основные виды регуляризации:**
- $L_{1}$-регуляризация (англ. lasso regression) через манхэттенское расстояние:
$ \displaystyle L_1 = \sum _{i}{(y_{i}-y(t_{i}))}^{2} + \lambda \sum _{i}{|a_{i}|}.$
- $L_{2}$-регуляризация или регуляризация Тихонова (в английской литературе — ridge regression или Tikhonov regularization):
$\displaystyle L_{2}=\sum _{i}{(y_{i}-y(t_{i}))}^{2}+\lambda \sum _{i}{a_{i}}^{2}$.

**Простыми словами:** Переобучение в большинстве случаев проявляется в том, что в получающихся многочленах слишком большие коэффициенты. Соответственно, и бороться с этим можно довольно естественным способом: нужно просто добавить в целевую функцию штраф, который бы наказывал модель за слишком большие коэффициенты.

Давайте посмотрим, как ведет себя решение для p = 8 и разных значений $\lambda$ при решении предыдущей задачи полиномиальной регрессии.
